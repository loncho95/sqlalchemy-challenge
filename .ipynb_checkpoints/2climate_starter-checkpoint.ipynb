```python
%matplotlib inline
from matplotlib import style
style.use('fivethirtyeight')
import matplotlib.pyplot as plt
```

```python
import numpy as np
import pandas as pd
import datetime as dt
from sqlalchemy import create_engine, inspect
from datetime import datetime, timedelta
from sqlalchemy import func
```

# Reflect Tables into SQLAlchemy ORM

```python
# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func
```

```python
# create engine to hawaii.sqlite
engine = create_engine("sqlite:///Resources/hawaii.sqlite")
```

```python
# reflect an existing database into a new model
Base = automap_base()
# reflect the tables
Base.prepare(autoload_with=engine)
```

```python
# View all of the classes that automap found
Base.classes.keys()
```

```python
# Save references to each table
measurement = Base.classes.measurement
station = Base.classes.station
```

```python
# Create our session (link) from Python to the DB
session = Session(engine)
```

# Exploratory Precipitation Analysis

```python
# Find the most recent date in the data set.
inspector = inspect(engine)
columns = inspector.get_columns('measurement')
for column in columns:
    print(column["name"], column["type"])

mostrecent_date = session.query(func.max(measurement.date)).scalar()
print(f"The most recent date in the sata set is {mostrecent_date}")
```

```python
# Design a query to retrieve the last 12 months of precipitation data and plot the results. 
# Starting from the most recent data point in the database. 
mostrecent_date = session.query(func.max(measurement.date)).scalar()
mostrecent_date = datetime.strptime(mostrecent_date, '%Y-%m-%d').date()

# Calculate the date one year from the last date in data set.
one_year_ago_v1 = mostrecent_date - timedelta(days=365)

# Perform a query to retrieve the data and precipitation scores
last_12_temp_v1 = session.query(measurement.date, measurement.prcp).filter(measurement.date > one_year_ago_v1).all()

# Save the query results as a Pandas DataFrame and set the index to the date column
last_12_temp_v1_df = pd.DataFrame(last_12_temp_v1, columns=['date','prcp']).set_index('date').dropna()


# Sort the dataframe by date
last_12_temp_v1_df.sort_index(inplace=True)


# Use Pandas Plotting with Matplotlib to plot the data
last_12_temp_v1_df.plot(figsize=(16,6))
plt.show()

```

```python
# Use Pandas to calcualte the summary statistics for the precipitation data
last_12_temp_v1_df.describe()
```

# Exploratory Station Analysis

```python
# Design a query to calculate the total number stations in the dataset
# confirm the variables within measurement class
# inspector = inspect(engine)
# columns = inspector.get_columns('measurement')
# for column in columns:
#     print(column["name"], column["type"])

number_stations = session.query(measurement).group_by(measurement.station).count()
print(f"the number of station is {number_stations}")
```

```python
# Design a query to find the most active stations (i.e. what stations have the most rows?)
# List the stations and the counts in descending order.
session.query(measurement.station, func.count(measurement.date)).group_by(measurement.station).\
    order_by(func.sum(measurement.station).desc()).all()
```

```python
# Using the most active station id from the previous query, calculate the lowest, highest, and average temperature
avg_temp = session.query(func.avg(measurement.tobs)).filter(measurement.station == 'USC00519281').scalar()
min_temp = session.query(func.min(measurement.tobs)).filter(measurement.station == 'USC00519281').scalar()
max_temp = session.query(func.max(measurement.tobs)).filter(measurement.station == 'USC00519281').scalar()


print(f"The average temperature is {avg_temp:.2f}, the lowest temperature is {min_temp:.2f}, and the highest temperature is {max_temp:.2f} ")
```

```python
# Using the most active station id
# Query the last 12 months of temperature observation data for this station and plot the results as a histogram

mostrecent_date_USC00519281 = session.query(func.max(measurement.date)).filter(measurement.station == 'USC00519281').scalar()
mostrecent_date_USC00519281 = datetime.strptime(mostrecent_date_USC00519281, '%Y-%m-%d').date()
# calculate the date 12 months ago
one_year_ago = mostrecent_date_USC00519281 - timedelta(days=365)
one_year_ago

#Obtaining the temperatures

last_12_temp = session.query(measurement.tobs).filter(measurement.station == 'USC00519281').filter(measurement.date > one_year_ago).all()
last_12_temp

last_12_temp_df = pd.DataFrame(last_12_temp, columns=['tobs'])

# Plot the histogram
plt.hist(last_12_temp_df['tobs'], bins=12, color='skyblue', alpha=0.8)

# Add labels and title
plt.xlabel('Temperature (F)')
plt.ylabel('Frequency')
plt.title('Temperature Observations for USC00519281\nLast 12 Months')

# Show the plot
plt.show()
```

# Close session
